# Sketchy lines homework

Имплементирован алгоритм визуализации sketchy lines по статье https://diglib.eg.org/bitstream/handle/10.2312/evs20201062/139-143.pdf?sequence=1&isAllowed=y

Все параметры из статьи поддерживаются

Можно рисовать как отдельные линии, так и line-based решётки с использованием шума Перлина для рандомизации параметров (реализация шума Перлина взята отсюда - https://github.com/ruslangrimov/perlin-noise-python-numpy). 

```python
import cairo
from IPython.display import Image, display
from io import BytesIO
from lines import sketchy_line, sketchy_grid

def disp(draw_func):
    surface = cairo.ImageSurface(cairo.FORMAT_ARGB32, 600, 600)
    ctx = cairo.Context(surface)
    draw_func(ctx, 600, 600)
    with BytesIO() as fileobj: 
        surface.write_to_png(fileobj)
        display(Image(fileobj.getvalue(), width=600))
```

## Sketchy lines demo

```python
@disp
def sketchy_line_demo(cr, width, height):
    cr.scale(width, height)
    sketchy_line(cr, [0.1, 0.2], [0.9, 0.2], 5, 0.017, [0.001, 0.005], 1, [1], -7)
    sketchy_line(cr, [0.1, 0.4], [0.9, 0.4], 7, 0.014, [0.0005, 0.005], 3, [1], 7)
    sketchy_line(cr, [0.1, 0.6], [0.9, 0.6], 13, 0.009, [0.001, 0.005], 3, [3, -1], 2)
    sketchy_line(cr, [0.1, 0.8], [0.9, 0.8], 15, 0.015, [0.0005, 0.003], 1, [1], -2)
```

## Sketchy grid demo

```python
@disp
def sketchy_grid_demo_1(cr, width, height):
    cr.scale(width, height)
    sketchy_grid(cr, 10, [3, 18], [0.004, 0.009], [0.0005, 0.005], [-7, 7])

@disp
def sketchy_grid_demo_2(cr, width, height):
    cr.scale(width, height)
    sketchy_grid(cr, 10, [3, 18], [0.004, 0.009], [0.0005, 0.005], [-7, 7])
```
